```python
# Import necessary libraries
from IPython.display import display, Javascript
from google.colab.output import eval_js
from base64 import b64decode
import cv2
import numpy as np
import matplotlib.pyplot as plt

def capture_image():
    """
    Capture an image from the webcam in Google Colab.
    """
    # JavaScript code to capture an image from the webcam
    js = """
    async function takePhoto() {
      const div = document.createElement('div'); // Create a div element
      const capture = document.createElement('button'); // Create a button element
      capture.textContent = 'Capture'; // Set the button text
      div.appendChild(capture); // Add the button to the div
      document.body.appendChild(div); // Add the div to the body

      const video = document.createElement('video');
      video.style.display = 'block'; // Make the video element visible
      const stream = await navigator.mediaDevices.getUserMedia({video: true});
      document.body.appendChild(video); // Add the video element to the body
      video.srcObject = stream;
      await video.play();

      // Resize the output to have width 160 px and height 120 px
      video.width = 160;
      video.height = 120;

      // Wait for the user to click the Capture button
      await new Promise((resolve) => capture.onclick = resolve);

      const canvas = document.createElement('canvas');
      canvas.width = video.videoWidth;
      canvas.height = video.videoHeight;
      canvas.getContext('2d').drawImage(video, 0, 0);

      stream.getTracks().forEach(track => track.stop());
      video.remove();

      // Return the image as a base64 encoded string
      return canvas.toDataURL('image/jpeg');
    }
    """

    # Execute the JavaScript to capture the image
    display(Javascript(js))
    data = eval_js('takePhoto()')

    # Decode the base64 image data
    header, encoded = data.split(",", 1)
    image_data = b64decode(encoded)
    np_img = np.frombuffer(image_data, dtype=np.uint8)

    # Convert the image to an OpenCV format
    img = cv2.imdecode(np_img, cv2.IMREAD_COLOR)
    return img

# Capture an image and display it
img = capture_image()
if img is not None:
    # Display the captured image using matplotlib
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.show()
else:
    print("No image captured.")








import spacy
from transformers import pipeline, AutoTokenizer, AutoModelForSequenceClassification
import json
import cv2
import pytesseract  # Tesseract OCR library

# Initialize models
sentiment_analyzer = pipeline("sentiment-analysis", model="distilbert-base-uncased-finetuned-sst-2-english")
nlp_ner = spacy.load("en_core_web_sm")  # Load SpaCy model for NER, assuming it’s fine-tuned for mental health
tokenizer = AutoTokenizer.from_pretrained("distilbert-base-uncased")
model_classifier = AutoModelForSequenceClassification.from_pretrained("distilbert-base-uncased", num_labels=5)

# Custom keywords for mental health-related concerns
custom_concerns = {
    "anxiety": ["anxious", "extremely anxious", "worried", "nervous"],
    "depression": ["depressed", "very low", "down", "hopeless"],
    "insomnia": ["trouble sleeping", "insomnia", "can't sleep", "sleepless"],
    "stress": ["stressed", "under pressure", "overwhelmed"]
}

def capture_text_from_camera():
    # Start video capture
    cap = cv2.VideoCapture(0)
    print("Press 's' to capture and process the image, or 'q' to quit.")

    while cap.isOpened():
        ret, frame = cap.read()
        if ret:
            cv2.imshow('Camera', frame)
            if cv2.waitKey(1) & 0xFF == ord('s'):
                # Save and process image on pressing 's'
                captured_text = process_image_with_ocr(frame)
                print(f"Captured Text: {captured_text}")
                cap.release()
                cv2.destroyAllWindows()
                return captured_text
            elif cv2.waitKey(1) & 0xFF == ord('q'):
                # Quit on pressing 'q'
                cap.release()
                cv2.destroyAllWindows()
                return None
        else:
            break

def process_image_with_ocr(image):
    # Convert image to grayscale
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    # Perform OCR on the image
    text = pytesseract.image_to_string(gray)
    return text.strip()

def extract_concerns(text, doc):
    extracted_concerns = [ent.text for ent in doc.ents if ent.label_ in ["ANXIETY", "DEPRESSION", "STRESS", "INSOMNIA"]]

    # Check for custom keyword matches
    for category, phrases in custom_concerns.items():
        for phrase in phrases:
            if phrase in text:
                extracted_concerns.append(phrase)

    return extracted_concerns

# Define the function
def determine_mental_health_issue(text, save_to_file=False):
    # Initialize the output structure
    output = {"polarity": None, "confidence": None, "concerns": []}

    # Step 1: Polarity Detection
    polarity_result = sentiment_analyzer(text)[0]
    output["polarity"] = polarity_result['label']
    output["confidence"] = polarity_result['score']
    print(f"Polarity: {output['polarity']} (Confidence: {output['confidence']})")

    # Step 2: Keyword Extraction (NER and Custom Keywords)
    doc = nlp_ner(text)
    extracted_concerns = extract_concerns(text, doc)

    # If no relevant concerns found
    if not extracted_concerns:
        return "No specific mental health concerns detected."

    # Step 3: Concern Classification
    for concern in extracted_concerns:
        inputs = tokenizer(concern, return_tensors="pt")
        outputs = model_classifier(**inputs)
        label = outputs.logits.argmax().item()

        # Map label to category name based on training labels
        categories = ["Anxiety", "Depression", "Stress", "Insomnia", "Eating Disorder"]
        classified_concern = categories[label]
        output["concerns"].append({"concern": concern, "category": classified_concern})

    # Structured final output
    print("\nFinal Determined Issues:")
    print(f"Polarity: {output['polarity']} (Confidence: {output['confidence']})")
    for item in output["concerns"]:
        print(f"Identified Concern: '{item['concern']}' classified as: {item['category']}")

    # Optionally save the output to a JSON file
    if save_to_file:
        with open("output.json", "w") as outfile:
            json.dump(output, outfile, indent=4)
        print("Output saved to output.json")

    return output

# Main function to capture text and analyze it
def main():
    print("Starting mental health concern analysis...")
    user_text = capture_text_from_camera()
    if user_text:
        result = determine_mental_health_issue(user_text, save_to_file=True)
        print("\nStructured Output:", result)
    else:
        print("No text captured from camera.")

# Run the main function
main()
import spacy
from transformers import pipeline, AutoTokenizer, AutoModelForSequenceClassification
import json

# Initialize models
sentiment_analyzer = pipeline("sentiment-analysis", model="distilbert-base-uncased-finetuned-sst-2-english")
nlp_ner = spacy.load("en_core_web_sm")  # Load SpaCy model for NER, assuming it’s fine-tuned for mental health
tokenizer = AutoTokenizer.from_pretrained("distilbert-base-uncased")
model_classifier = AutoModelForSequenceClassification.from_pretrained("distilbert-base-uncased", num_labels=5)

# Custom keywords for mental health-related concerns
custom_concerns = {
    "anxiety": ["anxious", "extremely anxious", "worried", "nervous"],
    "depression": ["depressed", "very low", "down", "hopeless"],
    "insomnia": ["trouble sleeping", "insomnia", "can't sleep", "sleepless"],
    "stress": ["stressed", "under pressure", "overwhelmed"]
}

def extract_concerns(text, doc):
    extracted_concerns = [ent.text for ent in doc.ents if ent.label_ in ["ANXIETY", "DEPRESSION", "STRESS", "INSOMNIA"]]

    # Check for custom keyword matches
    for category, phrases in custom_concerns.items():
        for phrase in phrases:
            if phrase in text:
                extracted_concerns.append(phrase)

    return extracted_concerns

# Define the function
def determine_mental_health_issue(text, save_to_file=False):
    # Initialize the output structure
    output = {"polarity": None, "confidence": None, "concerns": []}

    # Step 1: Polarity Detection
    polarity_result = sentiment_analyzer(text)[0]
    output["polarity"] = polarity_result['label']
    output["confidence"] = polarity_result['score']
    print(f"Polarity: {output['polarity']} (Confidence: {output['confidence']})")

    # Step 2: Keyword Extraction (NER and Custom Keywords)
    doc = nlp_ner(text)
    extracted_concerns = extract_concerns(text, doc)

    # If no relevant concerns found
    if not extracted_concerns:
        return "No specific mental health concerns detected."

    # Step 3: Concern Classification
    for concern in extracted_concerns:
        inputs = tokenizer(concern, return_tensors="pt")
        outputs = model_classifier(**inputs)
        label = outputs.logits.argmax().item()

        # Map label to category name based on training labels
        categories = ["Anxiety", "Depression", "Stress", "Insomnia", "Eating Disorder"]
        classified_concern = categories[label]
        output["concerns"].append({"concern": concern, "category": classified_concern})

    # Structured final output
    print("\nFinal Determined Issues:")
    print(f"Polarity: {output['polarity']} (Confidence: {output['confidence']})")
    for item in output["concerns"]:
        print(f"Identified Concern: '{item['concern']}' classified as: {item['category']}")

    # Optionally save the output to a JSON file
    if save_to_file:
        with open("output.json", "w") as outfile:
            json.dump(output, outfile, indent=4)
        print("Output saved to output.json")

    return output

# Example usage
user_input = "I feel extremely anxious and have trouble sleeping."
result = determine_mental_health_issue(user_input, save_to_file=True)
print("\nStructured Output:", result)
```

```python
!apt install tesseract-ocr -y
!pip install pytesseract fer transformers spacy

import cv2
import json
import spacy
from fer import FER
from transformers import pipeline, AutoTokenizer, AutoModelForSequenceClassification

# Initialize models for emotion analysis and classification
emotion_detector = FER(mtcnn=True)
sentiment_analyzer = pipeline("sentiment-analysis", model="distilbert-base-uncased-finetuned-sst-2-english")
nlp_ner = spacy.load("en_core_web_sm")
tokenizer = AutoTokenizer.from_pretrained("distilbert-base-uncased")
model_classifier = AutoModelForSequenceClassification.from_pretrained("distilbert-base-uncased", num_labels=5)

# Custom keywords for additional mental health concerns
custom_concerns = {
    "anxiety": ["anxious", "worried", "nervous"],
    "depression": ["depressed", "down", "hopeless"],
    "insomnia": ["trouble sleeping", "can't sleep", "sleepless"],
    "stress": ["stressed", "overwhelmed"],
    "happiness": ["happy", "joyful", "excited"],
    "satisfaction": ["content", "satisfied", "at ease"]
}

def extract_concerns(text):
    """
    Extract concerns based on text analysis using NER and keyword matching.
    """
    doc = nlp_ner(text)
    concerns = [ent.text for ent in doc.ents if ent.label_ in ["ANXIETY", "DEPRESSION", "STRESS", "INSOMNIA"]]
    for category, keywords in custom_concerns.items():
        concerns.extend([keyword for keyword in keywords if keyword in text])
    return concerns

def analyze_text_emotion(text):
    """
    Perform sentiment analysis on text input.
    """
    output = {"polarity": None, "confidence": None, "concerns": []}
    polarity_result = sentiment_analyzer(text)[0]
    output["polarity"] = polarity_result['label']
    output["confidence"] = polarity_result['score']

    concerns = extract_concerns(text)
    for concern in concerns:
        inputs = tokenizer(concern, return_tensors="pt")
        outputs = model_classifier(**inputs)
        label = outputs.logits.argmax().item()
        categories = ["Anxiety", "Depression", "Stress", "Insomnia", "Happiness", "Satisfaction"]
        classified_concern = categories[label] if label < len(categories) else "General"
        output["concerns"].append({"concern": concern, "category": classified_concern})

    return output

def track_emotions():
    """
    Track and display emotions in real-time using the webcam feed.
    """
    cap = cv2.VideoCapture(0)
    output_data = {"emotions": []}

    print("Press 'q' to quit.")
    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break

        # Detect emotion in the current frame
        emotion, score = emotion_detector.top_emotion(frame) or ("neutral", 0)
        if emotion:
            print(f"Detected Emotion: {emotion} with confidence: {score:.2f}")
            output_data["emotions"].append({"emotion": emotion, "confidence": score})
            cv2.putText(frame, f"{emotion} ({score:.2f})", (10, 40), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)

        # Display the resulting frame
        cv2.imshow("Emotion Tracking", frame)

        # Quit with 'q'
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

    # Save emotion data to JSON
    with open("emotion_data.json", "w") as file:
        json.dump(output_data, file, indent=4)
        print("Emotion data saved to emotion_data.json")

    # Release resources
    cap.release()
    cv2.destroyAllWindows()

# Example usage
track_emotions()
```
